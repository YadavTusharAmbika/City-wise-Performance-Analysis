```python
import pandas as pd

df = pd.read_csv("Dataset .csv")
df.head()
```

```python
city_counts = df["City"].value_counts()
city_counts.head()
```

```python
top_city = city_counts.idxmax()
top_city, city_counts.max()
```

```python
avg_rating_city = df.groupby("City")["Aggregate rating"].mean().sort_values(ascending=False)
avg_rating_city.head()
```

```python
best_city = avg_rating_city.idxmax()
best_city, avg_rating_city.max()
```

```python
# Conclusion:
# City with highest number of restaurants: top_city
# City with highest average rating: best_city
```
